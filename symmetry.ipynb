```python
#Import

from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
import numpy as np
from jupyter_jsmol import JsmolView
```

```python
with MPRester ("7dRNB3pyHFTcJ6LC") as m:
    #On crée notre stucture sur base du fichier MP
    Structure = m.get_structure_by_material_id("mp-2894")
```

```python
#Les variables dont on a besoin : 

Conv_Structure = SpacegroupAnalyzer(Structure).get_conventional_standard_structure()

SPanalyzer = SpacegroupAnalyzer(Conv_Structure)

Symm = SPanalyzer.get_symmetry_operations()
```

```python
#1ière Rotation l'atome de Sc en [0, 0, 0.5] par rapport a un axe d'ordre 2 :



print('\n',Symm[6].rotation_matrix,"[0,0,0.5]","+",Symm[6].translation_vector,"=",Symm[6].operate([0,0,0.5]))
```

```python
#2ième symétrie (de translation) sur l'atome de Sn situé en [0,0,0]
print('\n',Symm[24].rotation_matrix,"[0,0,0.0]","+",Symm[24].translation_vector,"=",Symm[24].operate([0,0,0.0]))
```
