```python
#### Import ####

from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
#### On récupère la structure de notre matériaux. ####

####On crée notre stucture sur base du fichier MP. ####
    
    
with MPRester ("7dRNB3pyHFTcJ6LC") as m:
   
    Structure = m.get_structure_by_material_id("mp-2894")
```

```python
#### On crée un XDR calculator pour notre diffractogramme. ####


xrd = XRDCalculator(1.54060)

#### On trouve le modèle de diffraction associé à notre cristal.####

Mod = xrd.get_pattern(Structure)

#### On imprime le diffractogramme.####

xrd.show_plot(Structure, annotate_peaks = True) 
```

```python
print('2*theta : ' + str(Mod.x[0]) + ' degrés' )
print('Intensité : ' + str(Mod.y[0]))
print('Indices h k l : ' + str(Mod.hkls[0]))
```
